# ***The Dataset is highly imbalanced, as the 0 & 1 category are varying too much***

```python
legit=data_set[data_set.Class==0]
fraud=data_set[data_set.Class==1]
```

```python
print(legit.shape,'\n',fraud.shape)
```

```python
data_set.groupby('Class').mean()
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

# ***Under-sampling the data, so that the dataset becomes balanced***

```python
legit_data=legit.sample(n=224)
legit.head()
```

```python
print(legit_data.shape,fraud.shape)
```

```python
final_dataset=pd.concat([legit_data,fraud],axis=0)
final_dataset.shape
```

```python
final_dataset.Class.value_counts()
```

```python
final_dataset.groupby('Class').mean()
```

```python
# The mean is approximately same as the previous original dataset, so the concatenated dataset is good for prediction.
```

```python
X=final_dataset.drop('Class',axis=1)
Y=final_dataset['Class']
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.2, stratify=Y,random_state=2)
```

```python
print(f'Shape of training features: {Xtrain.shape}')
print(f'Shape of testing features: {Xtest.shape}')
print(f'Shape of training targets: {Ytrain.shape}')
print(f'Shape of testing targets: {Ytest.shape}')
```

```python
model=LogisticRegression()
model.fit(Xtrain,Ytrain)
```

```python
Xtrain_prediction=model.predict(Xtrain)
Xtest_prediction=model.predict(Xtest)
```

```python
training_accuracy=accuracy_score(Ytrain,Xtrain_prediction)
print(training_accuracy)
```

```python
testing_accuracy=accuracy_score(Ytest,Xtest_prediction)
print(testing_accuracy)
```

```python
pickle.dump(model,open('model.pkl','wb'))
```
